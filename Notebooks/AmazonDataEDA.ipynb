```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('../Dataset/Superstore.csv',encoding='latin1')
#used encoding latin1 to properly read file.
df.head()
```

# Data Cleaning

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.duplicated().any()
```

```python
df.info()
```

```python
#Converting order dates and shipping dates to DateTime
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df.info()
```

# Exploratory Data Analysis

```python
df.describe()
```

```python
# 1) calculate the monthly sales
# 2) Identify which month has most and least sales

#We need to divide the date into month and year and add the columns
df['Order Month']=df['Order Date'].dt.month
df['Order Year']=df['Order Date'].dt.year

#show
df.head()
```

```python
sales_by_month=df.groupby('Order Month')['Sales'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.barplot(data=sales_by_month,y='Sales',x='Order Month', color="skyblue")
plt.title("Month Wise total Sales")
plt.show
```

```python
#since Feb and NNovember are the months, here are the total sales values
print(sales_by_month[sales_by_month['Order Month']==2])
print(sales_by_month[sales_by_month['Order Month']==11])
```

# Analyze sales based on product categories

```python
sales_by_category=df.groupby('Category')['Sales'].sum().reset_index().sort_values('Sales', ascending=False)
print(sales_by_category)
plt.figure(figsize=(6,5))
sns.barplot(data=sales_by_category,y='Sales',x='Category', color="skyblue", width=0.5)
plt.title("Category Wise total Sales")
plt.show()
```

# Sales analysis based on sub-categories

```python
# print(df['Sub-Category'].unique())
sales_by_sub_category=df.groupby('Sub-Category')['Sales'].sum().reset_index().sort_values('Sales', ascending=False)

print(sales_by_sub_category)
plt.figure(figsize=(11,8))
sns.barplot(data=sales_by_sub_category,x='Sales',y='Sub-Category', color="skyblue")
plt.title("Sub-Category Wise total Sales")
plt.show()
```

# Monthly profit from sales

```python
profit_by_month=df.groupby('Order Month')['Profit'].sum().reset_index()
print(profit_by_month)
plt.figure(figsize=(11,8))
sns.barplot(data=profit_by_month,x='Order Month',y='Profit', color="skyblue")
plt.title("Monthly Profits")
plt.show()
```

# Analyze the profit by category and sub-category

```python
profit_by_category=df.groupby('Category')['Profit'].sum().reset_index().sort_values('Profit',ascending=False)
print(profit_by_category)
plt.figure(figsize=(6,5))
sns.barplot(data=profit_by_category,x='Category',y='Profit', color="skyblue", width=0.5)
plt.title("Monthly Profits")
plt.show()
```

```python
profit_by_sub_category=df.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values('Profit',ascending=False)
print(profit_by_sub_category)
plt.figure(figsize=(11,8))
sns.barplot(data=profit_by_sub_category,y='Sub-Category',x='Profit', color="skyblue")
plt.title("Monthly Profits")
plt.show()
```

# Analyze the sales and profit by customer segment

```python
sales_profit_by_seg=df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index().round(2)
print(sales_profit_by_seg)
plt.figure(figsize=(6,5))
sns.barplot(data=sales_profit_by_seg,y='Sales',x='Segment', width=0.5, color="skyblue")
plt.ticklabel_format(style='plain', axis='y')
plt.title("Sales in Customer Segment")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.barplot(data=sales_profit_by_seg,y='Profit',x='Segment', width=0.4, color="navy")
plt.ticklabel_format(style='plain', axis='y')
plt.title("Profit from Customer Segment")
plt.show()
```

# Sales to profit ratio by Segment

```python
sales_profit_by_seg['Sales_to_Profit_Ratio'] = sales_profit_by_seg['Sales'] / sales_profit_by_seg['Profit']
print(sales_profit_by_seg)
plt.figure(figsize=(6,5))
sns.barplot(data=sales_profit_by_seg,y='Sales',x='Segment', width=0.5, color="skyblue", label="Sales")
sns.barplot(data=sales_profit_by_seg,y='Profit',x='Segment', width=0.4, color="navy", label="Profit")
plt.ticklabel_format(style='plain', axis='y')
plt.title("Sales VS Profits Ratio in Customer Segment")
plt.show()
```
